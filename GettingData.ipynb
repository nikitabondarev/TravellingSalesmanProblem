```python
import pandas as pd
import googlemaps
from itertools import tee
import numpy as np
```

```python
df_adress = df_adress.replace(np.nan, '', regex=True)
arr = df_adress.values.tolist()
```

```python
len(arr)
```

```python
google_ad = []
for i in range(0, len(arr)):
    google_ad.append(' '.join(arr[i]))
google_ad
```

```python
latitudes = []
longitudes = []   
```

```python
for i in range(0, len(google_ad)):
    coords = gmaps.geocode(google_ad[i])
    long = coords[0]['geometry']['location'] ['lng']
    lat = coords[0]['geometry']['location'] ['lat']
    longitudes.append(long)
    latitudes.append(lat)
    
```

```python
len(longitudes)
```

```python
#добавляем новые колонкм
df_adress['Latitude'] = latitudes
df_adress['Longitude'] = longitudes
df_adress
```

```python
#saving all data back to csv file
df_adress.to_csv('file_name.csv', encoding='utf-8')
```
